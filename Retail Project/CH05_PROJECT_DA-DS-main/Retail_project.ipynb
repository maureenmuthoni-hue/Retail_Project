```python
data_query = '''
select * from retail;
'''
retail_df= pd.read_sql(data_query,con=db_url)
```

```python
valid_df = retail_df[retail_df['Quantity'] >= 1]

```

```python
invalid_df = retail_df[retail_df['Quantity'] < 1]
```

```python
print(retail_df.columns.tolist())
```

```python
valid_price_df = retail_df[retail_df['UnitPrice'] >= 0]
```

```python
retail_df = retail_df[~retail_df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
null_columns = retail_df.columns[retail_df.isnull().any()]
print(null_columns.tolist())
```

```python
retail_df = retail_df.dropna(subset=['CustomerID'])
```

```python
retail_df = retail_df.drop_duplicates()
```

```python
retail_df['Revenue'] = retail_df['Quantity'] * retail_df['UnitPrice']
```

```python
retail_df['InvoiceDate'] = pd.to_datetime(retail_df['InvoiceDate'])
```

```python
retail_df['Year'] = retail_df['InvoiceDate'].dt.year
retail_df['Month'] = retail_df['InvoiceDate'].dt.month
retail_df['Day'] = retail_df['InvoiceDate'].dt.day
```

```python
retail_df.head()
```

```python
engine = create_engine("postgresql+psycopg2://postgres:12345@localhost:5432/retail")
```

```python
retail_df.to_sql( name='online_retail',con=engine,if_exists='replace',index=False)
```

```python
data_2011 = retail_df[retail_df['Year'] == 2011]
```

```python
monthly_revenue = data_2011.groupby('Month')['Revenue'].sum().reset_index()
monthly_revenue = monthly_revenue.sort_values('Month')
print(monthly_revenue)
```

```python
monthly_revenue['Revenue_Change'] = monthly_revenue['Revenue'].diff()
monthly_revenue['Percent_Change'] = monthly_revenue['Revenue'].pct_change() * 100
print(monthly_revenue)
```

```python
country_data = retail_df[retail_df['Country'] != 'United Kingdom']
```

```python
country_summary = country_data.groupby('Country').agg(
    Total_Revenue=pd.NamedAgg(column='Revenue', aggfunc='sum'),
    Total_Quantity=pd.NamedAgg(column='Quantity', aggfunc='sum')
).reset_index()
```

```python
country_summary = country_summary.sort_values(by='Total_Revenue', ascending=False)
```

```python
top_countries = country_summary.head(10)
print(top_countries)
```

```python
customer_summary = retail_df.groupby('CustomerID').agg(
    Total_Revenue=pd.NamedAgg(column='Revenue', aggfunc='sum')
).reset_index()
```

```python
customer_summary = customer_summary.sort_values(by='Total_Revenue', ascending=False)
```

```python
top_customers = customer_summary.head(10)
print(top_customers)
```

## Global Product Demand

```python
data = retail_df[retail_df['Country'] != 'United Kingdom']
```

```python
country_demand = data.groupby('Country').agg(
    Total_Quantity=pd.NamedAgg(column='Quantity', aggfunc='sum')
).reset_index()
```

```python
country_demand = country_demand.sort_values(by='Total_Quantity', ascending=False)
print(country_demand)
```
